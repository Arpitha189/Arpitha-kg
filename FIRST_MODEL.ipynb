```python
X=df[['AGE','GENDER','BLOOD PRESSURE','CHOLESTEROL LEVEL','SMOKING STATUS','DIABETES','OBESITY','PHYSICAL ACTIVITY','STRESS LEVEL']]
Y=df['RISK']
```

```python
X=data.drop(columns='RISK')
Y=data['RISK']
```

```python
Y
```

```python
#to split the test and train data
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=6)
```

```python
X_train
```

```python
#to find the length of the train data set
len(X_train)
```

```python
#to find the length of the test data set
len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
```

```python
#we are training the model 
clf.fit(X_train,Y_train)
```

```python
#X is the independent variable here we predicting the X test sample
clf.predict(X_test)
```

```python
#Y is the dependent variable 
Y_test
```

```python
#based on the X and Y test we are predicting the accuracy of the model
clf.score(X_test,Y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
dt_model_=DecisionTreeClassifier(criterion='gini',max_features=10,random_state=10)
dt_model_.fit(X_train,Y_train)
```

```python
Y_pred=dt_model_.predict(X_test)
accuracy=accuracy_score(Y_test,Y_pred)
print(f"Accuracy:{accuracy}")
```
